```python
conf = SparkConf().setMaster("local").setAppName("FriendsByAge")
sc = SparkContext.getOrCreate(conf=conf)
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
friend = sc.textFile('/content/drive/MyDrive/file/friends.csv')
```

```python
friend.collect()
```

```python
def parseLine(line):
    fields = line.split(',')
    age = int(fields[2])
    numFriends = int(fields[3])
    return (age, numFriends)
```

```python
rdd = friend.map(parseLine)
```

```python
totalsByAge = rdd.mapValues(lambda x: (x, 1)).reduceByKey(lambda x, y: (x[0] + y[0], x[1] + y[1]))
averagesByAge = totalsByAge.mapValues(lambda x: x[0] / x[1])
```

```python
results = averagesByAge.collect()
```

```python
for i in results:
  print(i)
```

```python
lines = sc.textFile('/content/drive/MyDrive/file/1800.csv')
```

```python
def parseLine(line):
    fields = line.split(',')
    stationID = fields[0]
    entryType = fields[2]
    temperature = float(fields[3]) * 0.1 * (9.0 / 5.0) + 32.0
    return (stationID, entryType, temperature)


parsedLines = lines.map(parseLine)
minTemps = parsedLines.filter(lambda x: "TMIN" in x[1])
```

```python
minTemps.take(5)
```

```python
stationTemps = minTemps.map(lambda x: (x[0], x[2]))
minTemps = stationTemps.reduceByKey(lambda x, y: min(x,y))
results = minTemps.collect();

for result in results:
    print(result[0] + "\t{:.2f}F".format(result[1]))
```

```python
maxTemps = parsedLines.filter(lambda x: "TMAX" in x[1])
```

```python
stationTemps = maxTemps.map(lambda x: (x[0], x[2]))
maxTemps = stationTemps.reduceByKey(lambda x, y: max(x,y))
results = maxTemps.collect();

for result in results:
    print(result[0] + "\t{:.2f}F".format(result[1]))
```

```python
from pyspark import SparkConf, SparkContext

conf = SparkConf().setMaster("local").setAppName("WordCount")
sc = SparkContext.getOrCreate(conf = conf)

```

```python
input = sc.textFile('/content/drive/MyDrive/file/book.txt')
words = input.flatMap(lambda x: x.split())
wordCounts = words.countByValue()

```

```python
wordCounts
```

```python
for word, count in wordCounts.items():
    cleanWord = word.encode('ascii', 'ignore')
    if (cleanWord):
        print(cleanWord.decode() + " " + str(count))
```

```python
amount = sc.textFile('/content/drive/MyDrive/file/amount.csv')
```

```python
def parse(line):
  fields = line.split(',')
  id = int(fields[0])
  amt = float(fields[2])
  return (id,amt)
```
